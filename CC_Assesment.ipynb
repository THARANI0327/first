# Time Series Analysis

```python
data['Invoice Date'] = pd.to_datetime(data['Invoice Date'])
print(data.dtypes)
```

```python
daily_sales = data.groupby('Invoice Date')['Sales Amount'].sum()
print(daily_sales)
```

```python
monthly_sales = daily_sales.resample('M').sum()
print(monthly_sales)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(daily_sales.index, daily_sales.values)
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.title('Daily Sales Trend')
plt.show()
```

```python
seasonal_sales = daily_sales.resample('QE').sum()
sns.lineplot(x=seasonal_sales.index, y=seasonal_sales.values)
plt.xlabel('Invoice Date')
plt.ylabel('Total Sales Amount')
plt.title('Quarterly Sales Trend')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales.index, monthly_sales.values)
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.title('Monthly Sales Trend')
plt.show()
```

```python
weekly_sales = daily_sales.resample('W').sum()
best_day = daily_sales.idxmax()
best_day_sales = daily_sales.max()
print(f"Best Performing Day: {best_day.date()} with sales of {best_day_sales:.2f}")
```

```python
best_week = weekly_sales.idxmax()
best_week_sales = weekly_sales.max()
print(f"Best Performing Week: {best_week.date()} with sales of {best_week_sales:.2f}")
```

```python
best_month = monthly_sales.idxmax()
best_month_sales = monthly_sales.max()
print(f"Best Performing Month: {best_month.date()} with sales of {best_month_sales:.2f}")
```
